*Caso 1: Relación entre sentimiento y aerolíneas*

Instrucciones:

- Carga de la base de datos: Importa la base de datos aerolineas.csv a tu entorno de trabajo.
- Revisión de datos: Realiza un análisis exploratorio inicial para verificar la estructura de los datos: Muestra las primeras filas del dataset. Verifica la existencia de datos faltantes y, si es necesario, realiza una limpieza básica.

Tablas de contingencia:

- Genera una tabla de contingencia que relacione las categorías de sentimientos (positivo, neutral y negativo) con las aerolíneas.

Prueba Chi Cuadrado:

- Realiza una prueba de Chi Cuadrado para determinar si existe una asociación significativa entre las variables.
- Presenta los valores de la estadística Chi Cuadrado, los grados de libertad y el p-valor.

Interpretación:

- Describe en un párrafo tus hallazgos en base al p-valor obtenido y concluye si existe o no una asociación significativa entre los sentimientos expresados en los mensajes y las aerolíneas mencionadas.

 

*Caso 2: Relación entre ingreso y variables de gasto*

Instrucciones:

- Carga de la base de datos: Importa la base de datos enigh202.csv a tu entorno de trabajo.
- Revisión de datos: Realiza un análisis exploratorio inicial:
- Inspecciona las primeras filas del dataset.
- Identifica las variables relevantes: 'ingreso corriente' y las variables de gasto `['alimentos','cereales','carnes','leche','huevo','bebidas','agua','energia','cuida_pers']`.
- Verifica la existencia de valores atípicos y ajusta los datos si es necesario.

Prueba de correlación:

- Calcula el coeficiente de correlación de Pearson entre el ingreso y cada una de las variables de gasto seleccionadas.
- Realiza una prueba de hipótesis para cada correlación obtenida (H0: No hay correlación significativa, H1: Hay correlación significativa).
- Presenta el valor del coeficiente de correlación, el p-valor asociado y concluye para cada prueba.

Visualización:

- Crea un gráfico de dispersión para el par de variables que presenten una mayor correlación significativa.

Interpretación: Describe en un párrafo tus hallazgos.

- ¿Qué variables presentan una relación significativa con el ingreso?

- ¿La dirección y magnitud de las correlaciones tienen sentido en el contexto del problema? Justifica tus respuestas.
